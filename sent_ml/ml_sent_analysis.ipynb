```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
clean_tweet_list = []
tweet_len_list = []
```

```python
def preprocess(raw_tweet, flag):
    # Remove Special Character
    letters_only = re.sub('[^a-zA-Z]',' ',raw_tweet)
    
    # Conver sentence into Lower Case
    letters_only = letters_only.lower()
    
    # Tokenize
    words = letters_only.split()
    
    #Remove Stop Words
    words = [w for w in words if not w in stopwords.words('english')]
    
    # Stemming
    
    if flag == 'Stem':
        words = [stemmer.stem(word) for word in words]
    else:
        words = [lemmatizer.lemmatize(word) for word in words]
        
    clean_tweet_list.append(' '.join(words))  # words = ['the', 'flight', 'got', 'delayed']  
    tweet_len_list.append(len(words))
```

```python
from tqdm import tqdm
tqdm.pandas()
```

```python
tweet_df['Text'].progress_apply(lambda x: preprocess(x,'Stem'))

tweet_df['Clean_tweet_stem'] = clean_tweet_list
tweet_df['Tweet_len'] = tweet_len_list
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(tweet_df, test_size =0.2, random_state = 0)
```

```python
# Saving the Clean_tweet_stem column from the dataframe into a list
train_clean_tweet = []
for tweet in train['Clean_tweet_stem']:
    train_clean_tweet.append(tweet)
    
test_clean_tweet = []
for tweet in test['Clean_tweet_stem']:
    test_clean_tweet.append(tweet)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(ngram_range = (1,3))

train_features = vectorizer.fit_transform(train_clean_tweet)# ['gvbhj','ghdsbcj','dsbchj']

test_features = vectorizer.transform(test_clean_tweet)
```

```python
classifiers = [
                LogisticRegression(),
                DecisionTreeClassifier(),
                KNeighborsClassifier(n_neighbors = 8)
]
```

```python
for classifier in classifiers:
    model = classifier.fit(train_features,train['sentiment'])
    pred = model.predict(test_features)
    accuracy = accuracy_score(pred,test['sentiment'])
    print("Accuracy: ",accuracy)
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
def preprocess_test(raw_tweet):
    # Remove Special Character
    letters_only = re.sub('[^a-zA-Z]',' ',raw_tweet)
    
    # Conver sentence into Lower Case
    letters_only = letters_only.lower()
    
    # Tokenize
    words = letters_only.split()
    
    #Remove Stop Words
    words = [w for w in words if not w in stopwords.words('english')]
    
    # Stemming
    
    words = [stemmer.stem(word) for word in words]
    
    clean_tweet = ' '.join(words)
    
    return clean_tweet
```

```python
def predict(tweet):
    
    vectorizer = load(open('Models/vect.pkl','rb'))
    model = load(open('Models/logistic.pkl','rb'))
    
    clean_tweet = preprocess_test(tweet)
    
    clean_tweet_vector = vectorizer.transform([clean_tweet])
    
    prediction = model.predict(clean_tweet_vector)
    
    return prediction
```

```python
tweet = input('Enter the tweet: ')

prediction = predict(tweet)

if prediction == 0:
    print('Negative Sentiment')
else:
    print('Positive Sentiment')

print(prediction)
```
