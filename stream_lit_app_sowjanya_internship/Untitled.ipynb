```python
df['latitude'] =  pd.to_numeric(df['latitude'],errors='coerce')

df['longitude'] = pd.to_numeric(df['longitude'],errors='coerce')
```

```python

df[['lat_radians_X','long_radians_X']] = (
    np.radians(df.loc[:,['latitude','longitude']])
)
```

```python
import numpy as np
from sklearn.neighbors import BallTree

tree = BallTree(df[['lat_radians_X','long_radians_X']], leaf_size=2)              
distances, indices = tree.query(df_copy[['lat_radians_X','long_radians_X']], k=5)                
print(indices)
```

```python
to_concat = []
for i, m in enumerate(indices):
    
    loc = df.iloc[i, :]['longitude']  # str of location of interest's name
    neighbors = df.iloc[m, :]['latitude']  # pd.Series of n+1 nearest neighbors' names

            # Remove self matches (based on name) or last match if no self match
    if loc in neighbors.values:
        
        mask = (neighbors != loc)
        neighbors = neighbors[mask]
               
    else:
        neighbors = neighbors[:-1]
                

            # df for visibility into distances and index in market list
    for _, neighbor in enumerate(neighbors):
         to_concat.append([loc, neighbor])

df_new = pd.DataFrame(to_concat, columns=['longitude',
                                              "latitude"
                                              ])

loc_map = df_new.groupby("longitude").agg({"latitude": lambda x: list(x)})['latitude'].to_dict()

df_new
```
